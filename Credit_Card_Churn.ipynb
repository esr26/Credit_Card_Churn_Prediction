```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(14,2, figsize=(50,100))

for idx, feature in enumerate(num_df.columns.values):
  sns.histplot(data=df, x=feature, kde=True, ax=axes[idx][0])
  sns.histplot(data=df, x=feature, hue='Gender', kde=True, ax=axes[idx][1])
fig.tight_layout()
fig.show()


  
```

```python
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.countplot(data=df, x='Gender', hue='Attrition_Flag')
plt.show()
print('Number of exisiting female customers:', df.loc[(df['Gender']=='F') & (df['Attrition_Flag']=='Existing Customer')]['Gender'].count())
print('Number of existing male customers:', df.loc[(df['Gender']=='M') & (df['Attrition_Flag']=='Existing Customer')]['Gender'].count())
print('*'*50)
print('Number of attrition female customers:', df.loc[(df['Gender']=='F') & (df['Attrition_Flag']=='Attrited Customer')]['Gender'].count())
print('Number of attrition male customers:', df.loc[(df['Gender']=='M') & (df['Attrition_Flag']=='Attrited Customer')]['Gender'].count())
```

Female customers tend to get churned more than male customers

```python
from scipy import stats

fig = plt.figure()
ax1 = fig.add_subplot(211)
prob = stats.probplot(df['Avg_Open_To_Buy'], dist=stats.norm, plot=ax1)
ax1.set_title('Skewed distribution against Gaussian Distribution')

ax2 = fig.add_subplot(212)
xt, a = stats.boxcox(df['Avg_Open_To_Buy'])
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Skewed distribution after Gaussian Distribution')
plt.show()
print(len(xt))
```

```python
num_df['Avg_Open_To_Buy'] = xt
sns.histplot(data=num_df, x='Avg_Open_To_Buy', kde=True)
plt.show()
```
